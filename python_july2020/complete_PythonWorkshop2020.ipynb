##Conditional Statements

if, else, elif

These statements help control the flow of your code.


Check variable RT:
* If response times are longer than or equal to 2000ms, feedback should be "too slow"
* If response times are shorter than or equal to 200ms, feedback should be "too fast"
* If response times are between 200 and 2000ms, feedback should be "good"



````
rt = 800

if (rt >= 2000):

  print("too slow")

elif (rt <= 200):

  print("too fast")

else:
  
  print("good")

````

##For Loops

Loops in python are a way to repeatedly
execute code inside the loop, until either a specific condition is met, or for a certain number of times.

````
# code that is inside the loop vs code that is outside the loop
# note that iteration starts with 0
# run the loop 5 times

for i in range(5):
  
  print("iteration", i) 
  print("this code is in the loop")
  

print("this is outside the loop")
````

For loops can be used to iterate through a list

````
mylist = ['this', 'is', 'an awesome', 'list']

# iterate using for loop and print the iteration and corresponding list element

for i in range (len(mylist)):
    
    print (i, mylist[i]) 
    
````

````
# another way

for i in mylist:

  print(i)
````

#### Question: Are tomatometer ratings higher than IMDB scores?

````
# Tomatometer ratings on a 10 scale

movies['Tomatometer10'] = movies['Tomatometer'] * 10

movies
````

````
# Are tomatometer ratings higher than IMDB scores?

movies['Tomatometer10'] > movies['IMDB_Score']
````

If we want to only display rows where tomatometer ratings higher than IMDB scores, we need to index the movies dataframe...

````
# only show rows where the tomatometer score is higher than the IMBD score
#movies['IMDB_Score'] < movies['Tomatometer10']

movies[movies['IMDB_Score'] < movies['Tomatometer10']]
````

#### Question: What is the mean of the difference between the Tomatometer and IMDB scores?

````
# find difference scores and the mean of the difference scores
# movies['IMDB_Score'] < movies['Tomatometer10']

movies['diff_score'] = movies['IMDB_Score'] - movies['Tomatometer10']
movies
````

````
# mean of the diff_score column

movies['diff_score'].mean()
````

````
# delete diff_score column

movies = movies.drop('diff_score', axis=1)
movies
````

#### Question: What is the average of the IMDB and Tomatometer scores for each movie?


````
# list of columns: ['IMDB_Score', 'Tomatometer10']

movies['mean'] = movies[['IMDB_Score', 'Tomatometer10']].mean(axis=1)
movies
````

````
# the lowest mean score

worst = movies['mean'].min()
worst
````

#### Group-By

**split-apply-combine**

We could also compare all groups within the data against each other, by using the split-apply-combine workflow. 

This splits data into groups, applies an operation on each group, and then combines the results into a table for display.

In pandas, we split into groups with the groupby function and then we apply an operation to the grouped data frame, e.g. .mean().


````
# group by drink

data.groupby('drink').mean()
````

````
# group by drink and condition

data.groupby(['drink','condition']).mean()
````

We can also easily count the number of observations in each group:

````
# to see how many data points are there in each group

data.groupby(['drink','condition']).size()
````

## Data Visualization using 'seaborn'



### Seaborn

* Seaborn is a library in python for statistical data visualization
* Seaborn is a Python data visualization library based on matplotlib 
* [More info](https://seaborn.pydata.org/#:~:text=statistical%20data%20visualization-,seaborn%3A%20statistical%20data%20visualization,can%20read%20the%20introductory%20notes.)

````
import seaborn as sns
````

#### **Plot 1** -- Dot Plot

Drink by Reaction Time dot plot showing reaction times from each trial

````
sns.swarmplot('drink', 'rt', data = data)
````

````
# settings below applies to all subsequent plots

sns.set(style='ticks', context='talk') 
````

````
# same plot with updated settings
sns.swarmplot('drink', 'rt', data=data)

# Removes the right and top lines around the plot
sns.despine() 
````

There are many other settings which can be used to update the plots and make them more presentable. 

https://seaborn.pydata.org/tutorial/aesthetics.html


####Remove Outlier

````
# create new column 'exclude'
# values in this column should be True for outliers

data['exclude'] = False

data.head()
````

````
# conditional value assignment using loc

data.loc[data.rt > 2000, 'exclude'] = True
data.loc[data.rt < 200, 'exclude'] = True
````

````
# check how many values are TRUE or FALSE in the exclude column

data['exclude'].value_counts()
````

````
# now select data without outliers

data_clean = data.loc[data.exclude == False, :]
data_clean
````

````
# plot again to see the data after the outliers were removed

sns.swarmplot('drink', 'rt', data=data_clean)
sns.despine() # Remove the right and top axes lines
  
````

####**Plot 2** -- Bar Plot

Plot a barplot of the clean data using seaborn. 
The bar plots should have reaction time (RT) split by drink and condition.

A good resource: https://seaborn.pydata.org/examples/index.html


````

sns.barplot('drink', 'rt', hue = 'condition', data=data_clean)
sns.despine() # Remove the right and top axes lines
````

####**Plot 3** -- Violin Plot

RTs split by Condition and Drink

````

sns.violinplot('drink', 'rt', hue = 'condition', data=data_clean)
sns.despine() # Remove the right and top axes lines
````

####**Plot 4** -- Combining Plots

Combining  Violin and Dot Plots


````

sns.violinplot('drink', 'rt', data=data_clean, inner = None, color = "lightgrey")
sns.swarmplot('drink', 'rt', data=data_clean, hue = 'condition', size = 4).set_title("Effect of drink on rt (ms)")
sns.despine()

````
